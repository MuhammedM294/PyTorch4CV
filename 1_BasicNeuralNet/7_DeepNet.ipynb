## 7. Building a Deeper Neural Network

All the previous models had have only one hidden layer. In this notebook, we will see the impact of having two hidden layers on the model perfomance.

```python
import torch 
import torch.nn as nn 
from torch.utils.data import Dataset, DataLoader
from torch.optim import SGD, Adam
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
device = 'cuda' if torch.cuda.is_available() else 'cpu' 
```

```python
from torchvision import datasets
data_folder = "Datasets" 
```

```python
fmnist = datasets.FashionMNIST(data_folder, download=True, train=True)
train_images = fmnist.data                                
train_targets = fmnist.targets
```

```python
validation_fmnist = datasets.FashionMNIST(data_folder, download=True, train=False)
validation_images = validation_fmnist.data                                
validation_targets = validation_fmnist.targets 
```

```python
def build_model(optimizer = Adam, lr = 1e-3):
    model = nn.Sequential(
        nn.Linear(28*28, 1000), 
        nn.ReLU(), 
        nn.Linear(1000,1000), 
        nn.ReLU(),
        nn.Linear(1000,10)
    ).to(device)
    
    loss_function = nn.CrossEntropyLoss()
    optimizer = optimizer(model.parameters(), lr = lr)
    
    return model , loss_function, optimizer
    
```

```python
from common_functions import get_data,train_with_validation,display_train_validation_results
```

```python
train_data_loader, validation_data_loader = get_data(32,train_images,train_targets,validation_images,validation_targets)
```

```python
model,loss_function, optimizer = build_model(optimizer = Adam , lr=1e-3)
```

```python
train_losses, train_accuracies, validation_losses, validation_accuracies = train_with_validation(25, train_data_loader,
                                                                                                 validation_data_loader,
                                                                                                 model,
                                                                                                 loss_function,
                                                                                                 optimizer)
```

```python
display_train_validation_results(25,train_losses, train_accuracies, validation_losses, validation_accuracies)
```
