```python
import torch 
import torch.nn as nn 
from torch.utils.data import Dataset, DataLoader
from torch.optim import SGD, Adam
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
device = 'cuda' if torch.cuda.is_available() else 'cpu'  
```

```python
from torchvision import datasets
data_folder = "Datasets"                         
fmnist = datasets.FashionMNIST(data_folder, download=True, train=True)
```

```python
train_images = fmnist.data                               
train_targets = fmnist.targets
```

```python
validation_fmnist = datasets.FashionMNIST(data_folder, download=True, train=False)
validation_images = validation_fmnist.data                                
validation_targets = validation_fmnist.targets 
```

```python
class FMNISTDataset(Dataset):
    def __init__(self,x,y):
        x = x.float()/255
        x = x.view(-1,1,28,28)            #reshape the input to (size,channel,height,width)
        self.x, self.y = x,y
    def __len__(self):
        return len(self.x)
    def __getitem__(self,indx):
        x,y = self.x[indx], self.y[indx]
        return x.to(device), y.to(device)
```

```python
def build_model(optimizer = Adam, lr = 1e-3):
    model = nn.Sequential(
        nn.Conv2d(1,64,kernel_size=3),
        nn.MaxPool2d(2),
        nn.ReLU(),
        nn.Conv2d(64,128,kernel_size=3),
        nn.MaxPool2d(2),
        nn.ReLU(),
        nn.Flatten(),
        nn.Linear(3200,256),
        nn.ReLU(),
        nn.Linear(256,10), 
    ).to(device)
    
    loss_function = nn.CrossEntropyLoss()
    optimizer = optimizer(model.parameters(), lr = lr)
    
    return model , loss_function, optimizer
    
```
